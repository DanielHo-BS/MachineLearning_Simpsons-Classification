```python
import numpy as np
import os
import PIL
import PIL.Image
import tensorflow as tf
import pathlib
```

```python
path_test="theSimpsons-test/test"
data_dir_test = pathlib.Path(path_test)
```

```python
image_count = len(list(data_dir_test.glob('*/*.jpg')))
print(image_count)
```

```python
batch_size = 16
img_height = 180
img_width = 180

test_ds = tf.keras.utils.image_dataset_from_directory(
  data_dir_test,
  image_size=(img_height, img_width),
  shuffle=False, seed=None,
  batch_size=batch_size)
```

```python
class_names = test_ds.class_names
print(class_names)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 10))
for images, labels in test_ds.take(1):
  for i in range(9):
    ax = plt.subplot(3, 3, i + 1)
    plt.imshow(images[i].numpy().astype("uint8"))
    plt.title(class_names[labels[i]])
    plt.axis("off")
```

```python
normalization_layer = tf.keras.layers.Rescaling(1./255)

normalized_ds = test_ds.map(lambda x, y: (normalization_layer(x), y))
image_batch, labels_batch = next(iter(normalized_ds))
first_image = image_batch[0]
# Notice the pixel values are now in `[0,1]`.
print(np.min(first_image), np.max(first_image))
```

```python
AUTOTUNE = tf.data.AUTOTUNE

test_ds = test_ds.cache().prefetch(buffer_size=AUTOTUNE)
```

```python
import keras
from keras.models import Sequential
from keras.models import Model
from keras.callbacks import ModelCheckpoint, LearningRateScheduler, EarlyStopping, ReduceLROnPlateau, TensorBoard
from keras import optimizers, losses, activations, models
from keras.layers import Convolution2D, Dense, Input, Flatten, Dropout, MaxPooling2D, BatchNormalization, GlobalAveragePooling2D, Concatenate,Dropout
from keras.applications.inception_v3 import InceptionV3
from keras.applications.resnet import ResNet50
```

```python
#載入訓練好的模型 
model = keras.models.load_model('lab-logs/models/Best-model-2.h5')
model.summary()
```

```python
predict=model.predict(test_ds)
```

```python
preds_classes = np.argmax(predict, axis=-1)
```

```python
preds_classes
```

```python
path="theSimpsons-train/train"
data_dir = pathlib.Path(path)

train_ds = tf.keras.utils.image_dataset_from_directory(
  data_dir,
  validation_split=0.2,
  subset="training",
  seed=123,
  image_size=(img_height, img_width),
  batch_size=batch_size)

class_names = train_ds.class_names
print(class_names)
```

```python
predict_name=[]
for i in preds_classes:
    predict_name.append(class_names[i])
```
